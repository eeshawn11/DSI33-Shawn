    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!

```python
# Code:
num_test = [66, 30, 40, 64]

def calc_mean(num):
    return sum(num) / len(num)

calc_mean(num_test)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:
def calc_std(num):
    # applying standard deviation formula
    num_mean = calc_mean(num)
    num_std = (sum([((i - num_mean) ** 2) for i in num]) / (len(num) - 1)) ** 0.5
    return f"{num_std:.6f}"

print(calc_std(num_test))
```

```python
# import SAT California 2019 dataset
sat_ca_import = pd.read_csv("../data/sat_2019_ca.csv")

sat_ca_import.tail()
```

```python
# rename columns to lower string
sat_ca_import.columns = sat_ca_import.columns.str.lower()
# replace * values with NaN to standardise
sat_ca_import.replace(to_replace="*", value=np.NaN, inplace=True)
# extract required columns only
sat_ca = sat_ca_import.loc[:, ["cds", "sname", "dname", "enroll12", "numtsttakr12", "pctbothbenchmark12"]]
# rename columns to make them easier to use
sat_ca.rename(columns={"cds": "code", "sname": "school", "dname": "district", "enroll12": "enrolment", "numtsttakr12": "test_taker", "pctbothbenchmark12": "pct_benchmark"}, inplace=True)
# omit rows which are not schools, i.e. code or school name are null
sat_ca = sat_ca[(sat_ca.code.notnull() == True) & (sat_ca.school.notnull() == True)]
# convert values to correct data type
sat_ca["code"] = sat_ca["code"].apply(int)
sat_ca["enrolment"] = sat_ca["enrolment"].map(int, na_action="ignore")
sat_ca["test_taker"] = sat_ca["test_taker"].map(int, na_action="ignore")
sat_ca["pct_benchmark"] = sat_ca["pct_benchmark"].apply(lambda x: float(x)/100)
sat_ca["pct_test"] = sat_ca["test_taker"] / sat_ca["enrolment"]

sat_ca.info()
```

```python
# import student poverty data from California Department of Education, 2018-19
frpm_import = pd.read_excel("../data/frpm1819.xlsx", header=1, sheet_name=1)
frpm_import.head()
```

```python
# rename columns
frpm_import.columns = frpm_import.columns.str.lower().str.replace(" ", "_").str.replace("\n", "")
# merge county, district and school codes
frpm_import["code"] = frpm_import[["county_code", "district_code", "school_code"]].astype(str).apply("".join, axis=1)
# extract required columns only
frpm = frpm_import[["code", "percent_(%)_eligible_frpm_(k-12)"]].copy()
frpm.rename(columns={"percent_(%)_eligible_frpm_(k-12)": "frpm_eligible"}, inplace=True)
frpm["code"] = frpm["code"].map(int)
print(frpm.info())
```

```python
# merge Californian high school information into single df
ca_sch_info = pd.merge(sat_ca, frpm, on="code")
ca_sch_info.info()

# save CSV file
ca_sch_info.to_csv("../data/output_ca_sch_info.csv")
```

## Conclusions and Recommendations

The SAT provides an objective metric in helping such students in the college-admissions process, as opposed to other aspects of applications such as essays or extracurricular activities, which are subjective components that higher income students will score better in.

While students from lower income backgrounds may require more support in achieving better SAT scores, it would be in the Department's interest in providing more resources to these students. This creates more opportunities for students to gain access to college education and drive upward economic mobility across the entire state.

### Recommendations:
- **Review of the School Facility Program (SFP)**<br>
The SFP has been assessed to be relatively regressive, as school facility funding has been shown to be higher in districts with higher median household income and lower in districts with higher concentrations of disadvantaged students. The condition and availablilty of facilities for students would directly impact their ability to study and prepare for tests, especially for lower income students that may not have access to a conducive facility in their homes.<br><br>

- **Increase in teaching staff and resources**<br>
The LCFF has been fully implemented this year, with districts now receiving additional funding based on their share of higher-need students (e.g. low income, English learner and/or foster youth). This better distribution of funds from the state could enable schools to increase their teacher to student ratios, so as to better provide extra support to students preparing for SAT.<br><br>
The department could also consider to implement programs that enable districts to cross share their tecahing resources. This could help to supplement districts or schools with more severe shortage of staff.<br><br>

- **Encourage more students to prepare and participate in the SAT.**<br>
Studies have shown that earning a college degree is one of the best ways to improve one's economic performance. Especially within the lower income districts, students should be encouraged to take full advantage of available fiscal support that helps to offset their exam fees and maximise their attempts at the SAT. Schools should also allocate more resources in helping students better prepare from an earlier grade.
